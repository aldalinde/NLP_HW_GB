**Translation from russian to english using pipline from transformers and https://huggingface.co/Helsinki-NLP/opus-mt-ru-en model**

```python
import transformers
```

```python
from transformers import BertTokenizer, AutoTokenizer, AutoModelForSeq2SeqLM, pipeline
```

```python
ru_en_tokenizer = AutoTokenizer.from_pretrained("Helsinki-NLP/opus-mt-ru-en")
ru_en_model = AutoModelForSeq2SeqLM.from_pretrained("Helsinki-NLP/opus-mt-ru-en")
```

```python
ru_en_translator = pipeline(model=ru_en_model, tokenizer=ru_en_tokenizer, task="translation_ru_to_en")
```

```python
ru_en_translator(["Погода сегодня отличная в отличии от прошлой недели.", "Вечером идёт сильный дождь обычно до утра!"])
```

**Translation from russian to armenian using pipline from transformers and https://huggingface.co/Helsinki-NLP/opus-mt-ru-hy model**

```python
ru_arm_tokenizer = AutoTokenizer.from_pretrained("Helsinki-NLP/opus-mt-ru-hy")
ru_arm_model = AutoModelForSeq2SeqLM.from_pretrained("Helsinki-NLP/opus-mt-ru-hy")
```

```python
ru_arm_translator = pipeline(model=ru_arm_model, tokenizer=ru_arm_tokenizer, task="translation_ru_to_hy")
```

```python
ru_arm_translator(["Погода сегодня отличная в отличии от прошлой недели.", "Вечером идёт сильный дождь обычно до утра!"])
```

The first sentence is correct in the first part ("Погода сегодня отличная") but the rest is just skipped. At the end is the right armenian sign for the end of the sentence.     
The second sentence is translated: "Вечером вечером идет очень сильный дождь, обычно утром идет дождь."

Not bad at all.
